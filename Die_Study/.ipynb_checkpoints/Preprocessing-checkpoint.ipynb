## We store each transformation step individually to give more options for testing, feel free to customise the structure.
- `image_directory`: Path to images
- `target_directory`: Path where the output will be created
- `exp1 to exp5`: transformation steps

```python
image_directory = "images/"#"image_directory/"
target_directory = "output/"#"directory_to_create_output/"
exp1 = target_directory+"grayscale/"
exp2 = target_directory+"Filter1/"
exp3 = target_directory+"Filter2/"
exp4 = target_directory+"Filter3/"
exp5= target_directory+"Filter4/"

folders = [exp1, exp2, exp3, exp4, exp5]

if not os.path.exists(target_directory):
    os.mkdir(target_directory)
for folder in folders:
    if not os.path.exists(folder):
        os.mkdir(folder)
```

# Step 1: Grayscale

grayscale_directory(image_directory, exp1)

```python
def apply_rof(tmp):
    target_img = tmp[0]
    output = tmp[1]
    !python image-processing/denoise.py {target_img} "rof" -o {output}

def apply_denoise(src, target):
    my_list = []
    for root, dirs, files in os.walk(src, topdown=False):
        for name in files:
            target_img = os.path.join(root, name)
            output = os.path.join(target, name)
            my_list.append((target_img, output))

    results = Parallel(n_jobs=12, prefer="threads")(delayed(apply_rof)(i) for i in my_list)
```
