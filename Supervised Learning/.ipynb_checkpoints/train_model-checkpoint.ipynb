```python
%%capture
import os
import numpy as np
import matplotlib.pyplot as plt
import PIL
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler
import tensorflow_addons as tfa
import pandas as pd
import random

from utils import plot_training_images, plot_cm, get_classification_report, plot_gradcam
```

## Defining the train and test set

```python
data_dir = "/home/bigdatalab/Downloads/original/train/"
test_dir = "/home/bigdatalab/Downloads/original/test/"
```

```python
folder = data_dir+"V/" #"define_a_folder_to_look_at"
```

```python
plot_training_images(folder)
```

## Preprocessing

```python
# chose a preprocessing function based on the architecture selected
prep_func = tf.keras.applications.vgg16.preprocess_input 
```

```python
# Settings
batch_size = 32
image_size = (224, 224)
class_set = sorted(os.listdir(data_dir)) # define classes based on folder structure
num_classes = len(class_set) # Number of classes
class_set
```

```python
%%capture
train_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  seed=123,
  image_size=image_size,
  batch_size=batch_size,
  label_mode='int',
  shuffle=True)
```

```python
%%capture
val_ds = tf.keras.utils.image_dataset_from_directory(
  test_dir ,
  seed=123,
  image_size=image_size,
  batch_size=batch_size,
  label_mode='int')
```

```python
AUTOTUNE = tf.data.AUTOTUNE

# apply prep func 
train_ds = train_ds.map(lambda x, y: (prep_func(x), y))
val_ds = val_ds.map(lambda x, y: (prep_func(x), y))

# shuffle train data
train_ds = train_ds.shuffle(len(train_ds)*batch_size,seed=42, reshuffle_each_iteration=True).cache().prefetch(AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

## Create Model

### Pre-trained model

```python
pre_trained_model = tf.keras.applications.VGG16(
                    include_top=False, weights="imagenet", input_tensor=None,
                    input_shape=(224,224,3), pooling="avg",
                    )
```

```python
pre_trained_model.summary()
```

```python
# freeze some layers
for layer in pre_trained_model.layers:
    if layer.name.startswith("block5_conv3"):#
        break
    else:
        layer.trainable=False
```

```python
trainable = 0
frozen = 0
for layer in pre_trained_model.layers:
    if layer.trainable:
        trainable += 1
    else:
        frozen += 1
print(trainable, "trainable layers")
print(frozen, "frozen layers")
```

## Add layers

```python
# add dense and dropout layer. Adjust dropout rate if the model is overfitting.
dropout= 0.25
```

```python
x = tf.keras.layers.Dense(1024, activation="relu")(pre_trained_model.output)
x = tf.keras.layers.Dropout(dropout*3)(x)
x = tf.keras.layers.Dense(1024, activation="relu")(x)
x = tf.keras.layers.Dropout(dropout*2)(x)
x = tf.keras.layers.Dense(512, activation="relu")(x)
x = tf.keras.layers.Dropout(dropout*2)(x)
x = tf.keras.layers.Dense(num_classes, activation="softmax", name="predictions")(x)

model = tf.keras.models.Model(inputs=pre_trained_model.input, outputs=x)
```

## Define optimizer, loss and additional functions

```python
# define early stop 

class Callback(tf.keras.callbacks.Callback):
    """
    This function will be called at the end of each epoch.
    In case the thresholds set are reached the training will stop.
    View here some other possibilities to interact with the training process: 
    https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/Callback
    https://www.tensorflow.org/guide/keras/writing_your_own_callbacks
    """
    def on_epoch_end(self, epoch, logs={}):
        if logs.get("val_loss") <= 0.4:
            print("val_loss reached <40%")
            self.model.stop_training=True
        elif logs.get("loss") < 0.25:
            print("loss reached <25%")
            self.model.stop_training=True
        elif  logs.get("accuracy") >= 0.93:
            print("accuracy reached >93%")
            self.model.stop_training=True
```

```python
earlystop = Callback()
```

```python
optimizer =keras.optimizers.SGD(learning_rate= 0.0001, momentum=0.9)
```

```python
loss = tf.keras.losses.SparseCategoricalCrossentropy()
```

```python
model.compile(optimizer=optimizer, loss= loss, metrics=["accuracy"])
```

## Train

```python
epoch = 100 
history = model.fit(train_ds, validation_data = val_ds, epochs = epoch, callbacks = [earlystop])
```
